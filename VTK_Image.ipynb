```python
from pyevtk.hl import imageToVTK 

import numpy as np  

# Dimensions 

nx, ny, nz = 4, 16, 2 

ncells = nx * ny * nz 

npoints = (nx + 1) * (ny + 1) * (nz + 1) 

# Variables 

pressure = np.random.rand(npoints).reshape( (nx+1, ny+1, nz+1)) 

temp = np.random.rand(ncells).reshape( (nx, ny, nz), order = 'C') 

imageToVTK("./image", pointData = {"pressure" : pressure}, cellData = {"temp" : temp} )
```
